```python
# ===============================
# 1️⃣ Import des bibliothèques
# ===============================
# Pour manipuler les données et tracer les graphiques
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
# ===============================
# 2️⃣ Charger le dataset
# ===============================
# On suppose que le fichier CSV est dans le dossier "dataset"
df = pd.read_csv("../dataset/etudiant.csv")  # ajuster le chemin si nécessaire

# Afficher un aperçu pour vérifier les données
print("Aperçu du dataset :")
print(df.head())
# ===============================
# 3️⃣ Visualiser la relation entre Heures d'étude et Note
# ===============================
# On trace un nuage de points pour voir la relation
plt.scatter(df['Heures_etude'], df['Note'], color='blue')
plt.xlabel("Heures d'étude par jour")
plt.ylabel("Note finale (/20)")
plt.title("Relation entre Heures d'étude et Note")
plt.show()
# ===============================
# 4️⃣ Préparer les données pour le modèle
# ===============================
# X = variable indépendante (heures d'étude)
# y = variable dépendante (note)
X = df[['Heures_etude']]
y = df['Note']
# ===============================
# 5️⃣ Créer et entraîner le modèle
# ===============================
# On utilise la régression linéaire
model = LinearRegression()
model.fit(X, y)

# Récupérer les coefficients de la droite
a = model.coef_[0]      # pente
b = model.intercept_    # ordonnée à l'origine
print(f"Modèle entraîné : Note = {a:.2f} * Heures_etude + {b:.2f}")
# ===============================
# 6️⃣ Prédictions et évaluation
# ===============================
y_pred = model.predict(X)

# Calculer MSE et R² pour voir la qualité du modèle
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Erreur quadratique moyenne (MSE) : {mse:.2f}")
print(f"Coefficient de détermination (R²) : {r2:.2f}")
# ===============================
# 7️⃣ Visualiser la droite de régression
# ===============================
plt.scatter(X, y, color='blue', label='Données réelles')
plt.plot(X, y_pred, color='red', label='Droite de régression')
plt.xlabel("Heures d'étude par jour")
plt.ylabel("Note finale (/20)")
plt.title(f"Régression linéaire : Note = {a:.2f}*Heures_etude + {b:.2f}")
plt.legend()
plt.show()
# ===============================
# 8️⃣ Estimer les notes pour 15 nouveaux cas
# ===============================
# On veut prédire pour 1 à 15 heures d'étude
nouveaux_cas = pd.DataFrame({'Heures_etude': np.arange(1, 16)})

# Prédire les notes
notes_predites = model.predict(nouveaux_cas)
nouveaux_cas['Note_predite'] = notes_predites

print("\nPrédictions pour 15 cas :")
print(nouveaux_cas)

# Visualiser sur un graphique
plt.scatter(X, y, color='blue', label='Données réelles')
plt.plot(nouveaux_cas['Heures_etude'], nouveaux_cas['Note_predite'], color='green', linestyle='--', label='Prévisions 15 cas')
plt.xlabel("Heures d'étude par jour")
plt.ylabel("Note finale (/20)")
plt.title("Régression linéaire – Prédictions pour 15 cas")
plt.legend()
plt.show()
# ===============================
# 9️⃣ Estimer les notes pour 30 valeurs proches
# ===============================
# On prend 30 points uniformes entre le min et le max des heures d'étude
heures_min = df['Heures_etude'].min()
heures_max = df['Heures_etude'].max()
heures_30 = np.linspace(heures_min, heures_max, 30)  # 30 valeurs uniformes

# Créer un DataFrame pour ces nouvelles heures
nouveaux_30 = pd.DataFrame({'Heures_etude': heures_30})

# Prédire les notes avec le modèle
nouveaux_30['Note_predite'] = model.predict(nouveaux_30)

# Afficher les prédictions
print("\nPrédictions pour 30 valeurs proches :")
print(nouveaux_30)

# Visualiser
plt.scatter(X, y, color='blue', label='Données réelles')
plt.plot(nouveaux_30['Heures_etude'], nouveaux_30['Note_predite'], color='orange', linestyle='--', label='Prévisions 30 valeurs')
plt.xlabel("Heures d'étude par jour")
plt.ylabel("Note finale (/20)")
plt.title("Régression linéaire – Prédictions pour 30 valeurs proches")
plt.legend()
plt.show()
# ===============================
# 🔹 Estimer la note pour une valeur spécifique d'heures (limité entre 0 et 20)
# ===============================

heures = float(input("Entrez le nombre d'heures d'étude : "))

# Créer un DataFrame pour le modèle
valeur = pd.DataFrame({'Heures_etude': [heures]})

# Prédire la note
note_estimee = model.predict(valeur)[0]

# Limiter la note entre 0 et 20
note_estimee = max(0, min(20, note_estimee))

print(f"Pour {heures} heures d'étude, la note estimée est : {note_estimee:.2f} / 20")
```
