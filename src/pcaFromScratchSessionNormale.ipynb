```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================================================
# --- 1. EXTENSION DU DATASET À 50 ÉTUDIANTS (SESSION RATTRAPAGE) ---
# =============================================================================

# Création d'un grand tableau NumPy contenant les données de tous les étudiants.
# Chaque ligne représente un étudiant, et chaque colonne correspond à une variable spécifique :
# [Heures d'étude, Heures de sommeil, Niveau de stress, % Participation, Heures de sport, Temps d'écran, Note Finale]
raw_data = np.array([
    # Données des 14 étudiants initiaux
    [5, 6, 6, 55, 2, 4, 11], [2, 4, 8, 20, 0, 8, 7], [6, 7, 4, 65, 3, 3, 13],
    [1, 5, 9, 15, 1, 6, 5],  [4, 6, 5, 50, 2, 4, 10], [3, 5, 7, 35, 1, 5, 8],
    [7, 8, 3, 75, 4, 2, 14], [2, 6, 7, 30, 0, 7, 6],  [5, 5, 5, 60, 2, 3, 11],
    [3, 4, 8, 40, 1, 6, 9],  [6, 6, 4, 70, 3, 3, 12], [1, 3, 9, 10, 0, 9, 4],
    [4, 7, 6, 45, 2, 5, 10], [2, 5, 8, 25, 1, 7, 6],
    # Ajout de profils supplémentaires pour atteindre un effectif plus grand
    [6, 7, 3, 68, 2, 3, 12], [2, 5, 8, 22, 0, 7, 6],  [7, 7, 4, 80, 4, 2, 15],
    [3, 6, 6, 38, 1, 5, 9],  [5, 6, 5, 58, 2, 4, 11], [1, 4, 9, 12, 0, 8, 4],
    [4, 5, 6, 48, 1, 4, 10], [2, 4, 7, 28, 0, 6, 7],  [6, 8, 3, 72, 3, 2, 13],
    [3, 5, 8, 33, 1, 5, 8],  [5, 7, 4, 62, 2, 3, 12], [0, 3, 10, 5, 0, 10, 3],
    [7, 6, 3, 85, 3, 1, 16], [3, 4, 7, 42, 1, 6, 9],  [4, 6, 5, 52, 2, 3, 10],
    [2, 5, 9, 18, 0, 8, 5],  [6, 6, 5, 66, 3, 4, 12], [1, 5, 8, 16, 0, 7, 6],
    [5, 8, 4, 60, 4, 2, 11], [3, 5, 7, 30, 1, 6, 8],  [7, 7, 2, 78, 3, 2, 14],
    [2, 3, 9, 21, 0, 9, 5],  [4, 6, 6, 47, 2, 4, 10], [3, 6, 7, 36, 1, 5, 8],
    [6, 7, 4, 69, 2, 3, 13], [1, 4, 8, 14, 0, 8, 4],  [5, 5, 6, 57, 1, 4, 11],
    [2, 5, 7, 26, 0, 7, 7],  [7, 8, 3, 82, 4, 1, 15], [3, 4, 8, 41, 1, 5, 9],
    [4, 7, 5, 50, 2, 3, 10], [2, 6, 8, 24, 0, 8, 6],  [6, 6, 3, 74, 3, 2, 13],
    [1, 3, 9, 11, 0, 9, 3],  [5, 6, 5, 61, 2, 4, 11], [3, 5, 6, 34, 1, 6, 8],
    [6, 7, 4, 71, 3, 3, 13], [2, 4, 8, 19, 0, 8, 6],  [5, 6, 5, 59, 2, 4, 11],
    [4, 5, 6, 46, 1, 5, 10], [7, 8, 2, 88, 4, 1, 16], [1, 4, 9, 13, 0, 9, 4],
    [3, 5, 7, 37, 1, 6, 8],  [6, 6, 4, 73, 3, 2, 13], [2, 5, 8, 23, 0, 7, 6],
    [5, 7, 4, 64, 2, 3, 12], [4, 6, 5, 49, 2, 4, 10], [3, 4, 8, 39, 1, 6, 9],
    [7, 7, 3, 81, 4, 2, 15], [1, 3, 10, 8, 0, 10, 3], [6, 8, 3, 76, 3, 2, 14],
    [2, 6, 7, 29, 0, 7, 7],  [5, 5, 5, 56, 2, 4, 11], [4, 7, 5, 53, 2, 3, 10],
    [3, 5, 6, 32, 1, 5, 8],  [7, 8, 2, 90, 4, 1, 17], [1, 4, 9, 9,  0, 9, 3],
    [6, 7, 4, 67, 3, 3, 12], [2, 5, 8, 17, 0, 8, 5],  [5, 6, 5, 63, 2, 4, 11],
    [4, 5, 6, 44, 1, 5, 9],  [3, 4, 7, 31, 1, 6, 8],  [7, 7, 3, 84, 4, 2, 15],
    [1, 3, 9, 7, 0, 10, 2],  [6, 8, 3, 79, 3, 2, 14], [2, 6, 8, 27, 0, 7, 6],
    [5, 5, 5, 54, 2, 4, 10], [4, 7, 4, 51, 2, 3, 10], [3, 5, 7, 35, 1, 5, 8],
    [7, 8, 2, 92, 4, 1, 17], [1, 4, 10, 6, 0, 10, 2], [6, 7, 4, 70, 3, 3, 13],
    [2, 5, 8, 20, 0, 8, 6],  [5, 6, 5, 60, 2, 4, 11], [4, 5, 6, 45, 1, 5, 9],
    [3, 4, 7, 33, 1, 6, 8],  [7, 7, 3, 83, 4, 2, 15], [1, 3, 9, 5, 0, 10, 2],
    [6, 8, 3, 77, 3, 2, 14], [2, 6, 8, 25, 0, 7, 6],  [5, 5, 5, 58, 2, 4, 11],
    [4, 7, 4, 55, 2, 3, 10], [3, 5, 7, 36, 1, 5, 8],  [7, 8, 2, 95, 4, 1, 18],
    [1, 4, 10, 4, 0, 10, 1], [6, 7, 4, 72, 3, 3, 13], [2, 5, 8, 18, 0, 8, 5],
    [5, 6, 5, 62, 2, 4, 11], [4, 5, 6, 43, 1, 5, 9],  [3, 4, 7, 30, 1, 6, 8],
    [7, 7, 3, 86, 4, 2, 16]
])

# Extraction des caractéristiques explicatives : on sélectionne toutes les lignes (:) et les 6 premières colonnes (:6)
X_data = raw_data[:, :6]

# Extraction de la variable cible : on récupère uniquement la dernière colonne indexée par 6 (la note de l'étudiant)
notes = raw_data[:, 6]

# Recherche de la valeur maximale pour chaque colonne (caractéristique) du tableau X_data
max_vals = X_data.max(axis=0)

# Normalisation des données par mise à l'échelle (division de chaque valeur par le maximum de sa colonne)
X_norm = X_data / max_vals
```

```python
# =============================================================================
# --- 2. CLASSE PCA EN MACHINE LEARNING ---
# =============================================================================

# Définition de la classe personnalisée pour exécuter l'Analyse en Composantes Principales (ACP)
class MyPCA:
    # Constructeur de la classe prenant en paramètre le nombre de dimensions souhaitées en sortie
    def __init__(self, n_components):
        self.n_components = n_components # Stockage du nombre de composantes principales à conserver
        self.components = None            # Initialisation de l'attribut destiné à stocker les axes principaux (vecteurs propres)
        self.mean = None                  # Initialisation de l'attribut destiné à stocker la moyenne des données d'origine
        self.eigenvalues = None           # Initialisation de l'attribut destiné à stocker les valeurs propres

    # Méthode pour ajuster le modèle aux données et calculer les composantes principales
    def fit(self, X):
        # 1. Calcul de la moyenne pour chaque colonne du jeu de données d'entrée
        self.mean = np.mean(X, axis=0)

        # Soustraction de la moyenne à chaque observation pour centrer les données autour de l'origine (0,0)
        X_centered = X - self.mean

        # 2. Calcul de la matrice de covariance à partir du tableau des données centrées transposé
        covariance_matrix = np.cov(X_centered.T)

        # 3. Calcul automatique des valeurs propres et des vecteurs propres de la matrice de covariance
        eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

        # 4. Transposition de la matrice des vecteurs propres pour avoir un vecteur par ligne
        eigenvectors = eigenvectors.T

        # Tri des indices des valeurs propres de la plus grande à la plus petite (ordre décroissant)
        idxs = np.argsort(eigenvalues)[::-1]

        # Réorganisation des valeurs propres selon l'ordre trié (du plus grand impact au plus petit)
        self.eigenvalues = eigenvalues[idxs]

        # Réorganisation parallèle des vecteurs propres correspondants
        eigenvectors = eigenvectors[idxs]

        # 5. Sélection et stockage des K premiers vecteurs propres (les axes principaux demandés)
        self.components = eigenvectors[0 : self.n_components]

    # Méthode pour projeter les données initiales dans le nouvel espace de dimension réduite
    def transform(self, X):
        # Centrage des nouvelles données en soustrayant la moyenne calculée lors de l'étape de fit
        X_centered = X - self.mean

        # Projection des données centrées sur les axes principaux via un produit matriciel (produit scalaire)
        return np.dot(X_centered, self.components.T)
```

```python
# =============================================================================
# --- 3. EXÉCUTION ET RÉDUCTION EN 2D ---
# =============================================================================

# Instanciation de l'objet PCA configuré pour projeter le dataset en 2 dimensions
pca = MyPCA(n_components=2)

# Entraînement du modèle ACP sur la matrice des données normalisées
pca.fit(X_norm)

# Application de la transformation pour réduire les 6 caractéristiques de départ en 2 composantes principales
X_projected = pca.transform(X_norm)

# Calcul de la somme totale de toutes les valeurs propres obtenues (représente la variance totale globale)
variance_totale = np.sum(pca.eigenvalues)

# Calcul du pourcentage de variance capturée par la toute première composante principale (PC1)
var_pc1 = (pca.eigenvalues[0] / variance_totale) * 100

# Calcul du pourcentage de variance capturée par la deuxième composante principale (PC2)
var_pc2 = (pca.eigenvalues[1] / variance_totale) * 100

# Affichage des résultats textuels de l'analyse dans la console
print("--- ANALYSE ACP TERMINÉE ---")
print(f"Variance capturée par PC1 : {var_pc1:.2f}%")
print(f"Variance capturée par PC2 : {var_pc2:.2f}%")
print(f"Total de variance conservée : {var_pc1 + var_pc2:.2f}%\n")
```
