```python
import pandas as pd
import random

# Creation d'une liste vide pour stocker les donnees des etudiants
liste_etudiants = []

# Boucle pour creer 100 etudiants
for i in range(100):
    # Generation aleatoire des heures d'etudes entre 0 et 8
    heures_etudes = round(random.uniform(0, 8), 2)
    # Generation aleatoire de la participation entre 0 et 100
    participation = round(random.uniform(0, 100), 2)
    # Calcul de la note selon les heures et la participation
    note = (heures_etudes * 1.8) + (participation * 0.08) + random.uniform(-1.5, 1.5)
    # Verification que la note reste entre 0 et 20
    if note < 0:
        note = 0
    # Verification que la note ne depasse pas 20
    if note > 20:
        note = 20
    # Arrondissement de la note
    note = round(note, 2)
    # Condition pour definir le resultat final
    if note >= 10:
        resultat = "reussite"
    # Sinon l'etudiant est en echec
    else:
        resultat = "echec"
    # Generation des notes des modules
    java = round(random.uniform(5, 20), 2)
    cpp = round(random.uniform(5, 20), 2)
    xml = round(random.uniform(5, 20), 2)
    python_note = round(random.uniform(5, 20), 2)
    uml = round(random.uniform(5, 20), 2)
    proba_stat = round(random.uniform(5, 20), 2)
    # Ajout des informations de l'etudiant dans la liste
    liste_etudiants.append([heures_etudes, participation, note, resultat, java, cpp, xml, python_note, uml, proba_stat])

# Creation des noms des colonnes
colonnes = ["heures_etudes", "participation", "note", "resultat", "Java", "C++", "XML", "Python", "UML", "Proba_Stat"]
# Creation du DataFrame
data = pd.DataFrame(liste_etudiants, columns=colonnes)
# Affichage des 5 premieres lignes
print(data.head())
# Sauvegarde du dataset dans un fichier CSV
data.to_csv("session_normale.csv", index=False)
# Message de confirmation
print("Dataset cree avec succes")
```
