```python
import importlib
import naiveBayesModel

# Force le rechargement du fichier pour prendre en compte les modifications
importlib.reload(naiveBayesModel)

# Importation de la classe du modèle
from naiveBayesModel import NaiveBayesStudentModel
```

```python
# Creation du modele
model = NaiveBayesStudentModel()

# Chargement du dataset
data = model.load_data("session_normale.csv")

# Selection uniquement des etudiants en echec pour le rattrapage
etudiants_echec = data[data["resultat"] == "echec"]

# Affichage du nombre d'etudiants concernés
print("Nombre etudiants en rattrapage :", len(etudiants_echec))

# Preparation des donnees (X: variables, y: cible)
X, y = model.prepare_data(etudiants_echec)

# Entrainement du modele sur les donnees de rattrapage
model.train(X, y)
```

```python
# Lecture des heures d'études au clavier
heures = float(input("Entrer heures d'etudes : "))

# Lecture de la participation au clavier
participation = float(input("Entrer participation (%) : "))
```

```python
# Affichage du titre
print("===== SESSION RATTRAPAGE =====")
# Prediction via le modèle Naive Bayes
prediction, prob = model.predict_student(heures, participation)

# Règle métier stricte (Condition pour determiner le resultat final)
if heures >= 5 and participation >= 50:
    resultat_final = "reussite"
else:
    resultat_final = "echec_definitif"

# Affichage des résultats et des probabilités calculées
print("Prediction Naive Bayes :", prediction)
print("Resultat Final :", resultat_final)
print("Probabilites :", prob)

# Verdict final affiché à l'écran
if resultat_final == "reussite":
    print("Etudiant a reussi le rattrapage")
else:
    print("Etudiant en echec definitif")
```
