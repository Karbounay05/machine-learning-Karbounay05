```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import pandas as pd
from naiveBayesModel import NaiveBayesStudentModel
```

```python
# Creation du modele
model = NaiveBayesStudentModel()

# Chargement des donnees
data = model.load_data("session_normale.csv")

# Affichage des donnees
print(data.head())

# Preparation des donnees
X, y = model.prepare_data(data)

# Entrainement du modele
model.train(X, y)
```

```python
# Lecture des heures d'etudes
heures = float(input("Entrer heures d'etudes : "))

# Lecture de la participation
participation = float(input("Entrer participation (%) : "))
```

```python
# Affichage du titre
print("===== TEST SESSION NORMALE =====")

# Prediction du resultat (en passant les deux variables comme attendu)
prediction, prob = model.predict_student(heures, participation)

# Affichage du resultat
print("Resultat :", prediction)

# Affichage des probabilites
print("Probabilites :", prob)

# Condition si l'etudiant reussit
if prediction == "reussite":
    print("Etudiant valide la session normale")
# Sinon il passe au rattrapage
else:
    print("Etudiant va au rattrapage")
```
