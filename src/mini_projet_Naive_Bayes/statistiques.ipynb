```python
import pandas as pd

file_path = "session_normale.csv"  # Définition du chemin vers le fichier CSV
data = pd.read_csv(file_path)  # Chargement du fichier CSV dans un DataFrame pandas
```

```python
print("===== APERCU DU DATASET =====")  # Affichage d'un titre pour l'aperçu
print(data.head())  # Affichage des 5 premières lignes du DataFrame
print("===== INFORMATIONS =====")  # Affichage d'un titre pour les informations
print(data.info())  # Affichage des types de données, colonnes et valeurs manquantes
```

```python
# Création d'une liste contenant uniquement les noms des colonnes numériques
colonnes_numeriques = ["heures_etudes", "participation", "note", "Java", "C++", "XML", "Python", "UML", "Proba_Stat"]
```

```python
print("===== MOYENNE =====")  # Titre pour la moyenne
print(data[colonnes_numeriques].mean())  # Calcul et affichage de la moyenne pour chaque colonne numérique
```

```python
print("===== MEDIANE =====")  # Titre pour la médiane
print(data[colonnes_numeriques].median())  # Calcul et affichage de la valeur médiane de chaque colonne
```

```python
print("===== MIN =====")  # Titre pour le minimum
print(data[colonnes_numeriques].min())  # Recherche et affichage de la valeur minimale de chaque colonne
```

```python
print("===== MAX =====")  # Titre pour le maximum
print(data[colonnes_numeriques].max())  # Recherche et affichage de la valeur maximale de chaque colonne
```

```python
print("===== VARIANCE =====")  # Titre pour la variance
print(data[colonnes_numeriques].var())  # Calcul et affichage de la variance (dispersion des données)
```

```python
print("===== ECART TYPE =====")  # Titre pour l'écart-type
print(data[colonnes_numeriques].std())  # Calcul et affichage de l'écart-type (racine carrée de la variance)
```

```python
print("===== COVARIANCE =====")  # Titre pour la covariance
print(data[colonnes_numeriques].cov())  # Calcul de la matrice de covariance entre les variables
```

```python
print("===== CORRELATION =====")  # Titre pour la corrélation
print(data[colonnes_numeriques].corr())  # Calcul de la matrice de corrélation (liens linéaires entre -1 et 1)
```

```python
print("===== DESCRIPTION COMPLETE =====")  # Titre pour la description générale
print(data[colonnes_numeriques].describe())  # Génération d'un tableau regroupant les principales stats (count, mean, std, min, max, quartiles)
```
